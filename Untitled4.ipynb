# **Importing Required Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,precision_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline


```

# **Data Collection**

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully")
```

# **Data Preprocessing**

```python
data
```

```python
data.corr()
```

```python
data.duplicated()
```

# **Data Preparation**

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

# **Splitting the Data into Training and Testing Set**

```python

X_train, X_test, y_train, y_test = train_test_split(X, y,
                            test_size=0.2, random_state=0)
```

# **Training model using Linear Regression Algorithm**

```python
reg= LinearRegression()
```

```python
reg.fit(X_train, y_train)

print("Training complete.")
```

#**Plotting Datapoints**

```python

plt.figure(figsize=(20,6))
data.plot(x='Hours', y='Scores', style='*',color="orange",markersize=20)
plt.title('Hours Studied vs PercentageObtained')
plt.xlabel('Hours Studied')
plt.ylabel('PercentageObtained')
plt.show()
```

# **Plotting the regression line**



```python

line = reg.coef_*X+reg.intercept_

# Plotting for the test data
plt.scatter(X, y, color="orange", label='Actual Data', marker='*',s=95)


plt.plot(X, line,color="green");
plt.show()
```



# **Predicting the values for testing data**





```python
print(X_test) # Testing data - In Hours
y_pred = reg.predict(X_test) # Predicting the scores
```

# **Predicting Score of a Strudent who studies 9.25 hrs/day**

```python
hours_of_study = 9.25
predicted_score = reg.predict([[hours_of_study]])
predicted_score
```

## **Comparing Actual vs Predicted Values**



```python

df = pd.DataFrame({'Actual Values': y_test, 'Predicted values': y_pred})
df
```

# Evaluating of Model

```python
from sklearn import metrics
print('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test, y_pred))
from sklearn.metrics import r2_score,precision_score

# Assuming y_test and y_pred are defined
r2 = r2_score(y_test, y_pred)

print(f'R-squared: {r2}')




```
